```python
pdf = pd.read_excel('../../data/ucdp-term-acd-3-2021.xlsx', index_col=0)
pdf.head()
```

```python
gdf = pdf[pdf['territory_name'] == 'Palestine']
gdf.head()
```

```python
dct = []
def add_opp(x_in):
    x = x_in.split(', ')
    for i in x:
        dct.append(i)
    return x_in
pdf['location'] = pdf['location'].apply(add_opp)
```

```python
pdf['location'].value_counts()
```

```python
l = [1] * len(dct)

def chk(x):
    dct = {
        "Cambodia (Kampuchea)": "Cambodia",
        "DR Congo (Zaire)": "Democratic Republic of Congo",
        "Hyderabad": "India",
        "Madagascar (Malagasy)": "Madagascar",
        "Russia (Soviet Union)": "Russia",
        "Serbia (Yugoslavia)": "Serbia",
        "South Vietnam": "Vietnam",
        "Vietnam (North Vietnam)": "Vietnam",
        "South Yemen": "Yemen",
        "Yemen (North Yemen)": "Yemen",
    }
    if x in dct:
        return dct[x]
    return x

dfd = pd.DataFrame({'location': dct, 'count': l})
dfd['location'] = dfd['location'].apply(chk)

dfd.to_csv('generated.csv', index=False)
```

```python
ltc = list(gdf['side_a'])
ltc.extend(list(gdf['side_b']))
ltc
```

```python
ltc_new = []
for i in ltc:
    ltc_new.extend(i.split(', '))

len(ltc_new), len(ltc)
ltc = ltc_new
del ltc_new
```

```python
gdf['side_a'].value_counts()
```

```python
def get_year(x):
    return int(str(x)[:4])

gdf['Year'] = gdf['start_date2'].apply(get_year)

gdf['Year'].value_counts()
```

```python
def divide_actors(x):
    return x.split(', ')

gdf['actors_apart'] = gdf['side_a'].apply(divide_actors)
gdf['actors_bpart'] = gdf['side_b'].apply(divide_actors)

gdf['actors'] = gdf.apply(lambda x: x['actors_apart'] + x['actors_bpart'], axis=1)
```

```python
tdf = gdf.explode('actors')
```

```python
tdf = tdf[['Year', 'actors']]
```

```python
tdf = tdf[tdf['actors'] != 'Government of Israel']
```

```python
def year_bin(x):
    if x < 2000: return '1950-1999'
    if x < 2010: return '2000-2009'
    if x < 2025: return '2010-2025'

tdf['Year_bin'] = tdf['Year'].apply(year_bin)
tdf['Year_bin'].value_counts()
tdf[tdf['Year_bin'] == '1950-1999'].to_csv('1950-1999.csv', index=False)
```

```python
tdf.groupby(['actors']).size().to_csv('actors_grouped.csv')
```
