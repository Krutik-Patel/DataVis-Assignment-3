```python
df = pd.read_csv("./country_latLong.csv")
```

```python
merged_df = pd.merge(left=data, right=df, on='Country')
```

```python
merged_df["Country"].unique()
```

```python
merged_df = merged_df.loc[merged_df['Country'] == "Ukraine"]
```

```python
merged_df = merged_df.loc[merged_df['Year'] == 2022]
```

```python
merged_df.drop_duplicates(inplace=True)
merged_df.shape
```

```python
merged_df.head()
```

```python
# merged_df.to_csv("merged_w_latLong.csv", index=False)
```

```python
merged_df.drop(axis=1, columns=['Latitude', 'Longitude'], inplace=True)
```

```python
merged_df.index = range(len(merged_df["grouped_rows"]))
```

```python
for x in range(len(merged_df["grouped_rows"])):
    if merged_df["grouped_rows"][x][-2] != "}":
        merged_df["grouped_rows"][x] = merged_df["grouped_rows"][x] + "}]"
```

```python
for x in range(len(merged_df["grouped_rows"])):
    merged_df["grouped_rows"][x] = merged_df["grouped_rows"][x].replace("'", '')
    merged_df["grouped_rows"][x] = merged_df["grouped_rows"][x].replace(":", '')
    merged_df["grouped_rows"][x] = merged_df["grouped_rows"][x].replace(",", '')
```

```python
data0 = []
i = 1
for k in range(len(merged_df)):
    l0 = [x.replace("{", "") for x in merged_df.iloc[0]["grouped_rows"][1:-1].split("}")]
    for _ in range(len(l0)):
        l0[_] = l0[_].strip()
    for x in l0:
        l = x.split()
        # print(l)
        try:
            data0.append({
                "id": l[l.index("id") + 1],
                "latitude": l[l.index("latitude") + 1],
                "longitude": l[l.index("longitude") + 1],
                "count": l[l.index("best") + 1],
                # "count": merged_df[]["Total health worker arrested "] + merged_df["Total health worker injured"] + merged_df["Total health worker kidnapped"] + merged_df["Total health worker killed"],
                "Country": l[l.index("country") + 1]
            })
        except Exception as e:
            print(i, e)
            i += 1
    
```
